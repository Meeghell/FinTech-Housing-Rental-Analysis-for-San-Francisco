### Step 1: Use the `groupby` function to group the data by year. Aggregate the results by the `mean` of the groups.

```python
# Create a numerical aggregation that groups the data by the year and then averages the results.

# Select only the numeric columns for aggregation
housing_units_by_year = sfo_data_df.groupby('year')['housing_units'].mean()


# Review the DataFrame
# YOUR CODE HERE
housing_units_by_year
```

### Step 2: Use the `hvplot` function to plot the `housing_units_by_year` DataFrame as a bar chart. Make the x-axis represent the `year` and the y-axis represent the `housing_units`.

### Step 3: Style and format the line plot to ensure a professionally styled visualization.

```python
# Create a visual aggregation explore the housing units by year
# YOUR CODE HERE

# Calculate the minimum and maximum values for 'housing_units'
min_housing_units = housing_units_by_year.min()
max_housing_units = housing_units_by_year.max()

# Set the y-axis limits with a buffer based on the desired price level difference
buffer = 2500  # This is the desired price level difference
ylim = (min_housing_units - buffer, max_housing_units + buffer)

housing_units_plot = housing_units_by_year.hvplot.bar(
    x='year',
    y='housing_units',
    xlabel='Year',
    ylabel='Average Housing Units',
    title='Average Housing Units in San Francisco Per Year',
    yformatter='%.0f',
    height=400,
    width=600,
    ylim=ylim
).opts(
    color='skyblue',
    alpha=0.8,
    line_width=2,
    fontsize={'title': '13pt', 'labels': '11pt', 'xticks': '10pt', 'yticks': '10pt'},
    toolbar='above',
    yaxis='left'
)
housing_units_plot
```

### Step 1: Read the `neighborhood_coordinates.csv` file from the `Resources` folder into the notebook, and create a DataFrame named `neighborhood_locations_df`. Be sure to set the `index_col` of the DataFrame as “Neighborhood”.

```python
# Load neighborhoods coordinates data
neighborhood_locations_df = pd.read_csv(
    Path("Resources", "neighborhoods_coordinates.csv"),
    index_col="Neighborhood"
)
 # YOUR CODE HERE

# Review the DataFrame
# YOUR CODE HERE
print(neighborhood_locations_df.head())
```

### Step 2: Using the original `sfo_data_df` Dataframe, create a DataFrame named `all_neighborhood_info_df` that groups the data by neighborhood. Aggregate the results by the `mean` of the group.

```python
# Calculate the mean values for each neighborhood
all_neighborhood_info_df = sfo_data_df.groupby('neighborhood').mean() # YOUR CODE HERE

# Review the resulting DataFrame
# YOUR CODE HERE
print(all_neighborhood_info_df.head())
```

### Step 3: Review the two code cells that concatenate the `neighborhood_locations_df` DataFrame with the `all_neighborhood_info_df` DataFrame. 

Note that the first cell uses the [Pandas concat function](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html) to create a DataFrame named `all_neighborhoods_df`. 

The second cell cleans the data and sets the “Neighborhood” column. 

Be sure to run these cells to create the `all_neighborhoods_df` DataFrame, which you’ll need to create the geospatial visualization.

```python
# Using the Pandas `concat` function, join the 
# neighborhood_locations_df and the all_neighborhood_info_df DataFrame
# The axis of the concatenation is "columns".
# The concat function will automatially combine columns with
# identical information, while keeping the additional columns.
all_neighborhoods_df = pd.concat(
    [neighborhood_locations_df, all_neighborhood_info_df], 
    axis="columns",
    sort=False
)

# Review the resulting DataFrame
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

```python
# Call the dropna function to remove any neighborhoods that do not have data
all_neighborhoods_df = all_neighborhoods_df.reset_index().dropna()

# Rename the "index" column as "Neighborhood" for use in the Visualization
all_neighborhoods_df = all_neighborhoods_df.rename(columns={"index": "Neighborhood"})

# Review the resulting DataFrame
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

### Step 4: Using hvPlot with GeoViews enabled, create a `points` plot for the `all_neighborhoods_df` DataFrame. Be sure to do the following:

* Set the `geo` parameter to True.
* Set the `size` parameter to “sale_price_sqr_foot”.
* Set the `color` parameter to “gross_rent”.
* Set the `frame_width` parameter to 700.
* Set the `frame_height` parameter to 500.
* Include a descriptive title.

```python
# Create a plot to analyze neighborhood info
# YOUR CODE HERE


# Create a points plot using hvPlot with GeoViews enabled
points_plot = all_neighborhoods_df.hvplot.points(
    'Lon', 'Lat',  # Assuming your DataFrame has 'Lon' and 'Lat' columns for longitude and latitude
    geo=True,
    size='sale_price_sqr_foot',
    color='gross_rent',
    tiles='OSM',
    frame_width=700,
    frame_height=500,
    title='San Francisco Neighborhoods - Sale Price per Square Foot and Gross Rent',
    hover_cols=['Neighborhood']
)

# Display the plot
points_plot
```

## Compose Your Data Story

Based on the visualizations that you have created, compose a data story that synthesizes your analysis by answering the following questions:

**Question:**  How does the trend in rental income growth compare to the trend in sales prices? Does this same trend hold true for all the neighborhoods across San Francisco?


**Question:** What insights can you share with your company about the potential one-click, buy-and-rent strategy that they're pursuing? Do neighborhoods exist that you would suggest for investment, and why?


```python
# Answer: 

# Executing analysis on trends in rental income growth and sales prices
rental_income_trend = sfo_data_df.groupby('year')['gross_rent'].mean()
sales_price_trend = sfo_data_df.groupby('year')['sale_price_sqr_foot'].mean()

# Comparing trends across different neighborhoods
neighborhood_trends = sfo_data_df.groupby(['year', 'neighborhood']).mean()[['sale_price_sqr_foot', 'gross_rent']]

# Calculating the year-over-year growth for rental income and sales price
neighborhood_trends['rental_income_growth'] = neighborhood_trends.groupby('neighborhood')['gross_rent'].pct_change()
neighborhood_trends['sales_price_growth'] = neighborhood_trends.groupby('neighborhood')['sale_price_sqr_foot'].pct_change()

# Averaging the growth rates across all neighborhoods for each year
average_growth_rates = neighborhood_trends.groupby('year')[['rental_income_growth', 'sales_price_growth']].mean()

# Providing insights and recommendations
print("Data Story Insights and Recommendations:")

# How does the trend in rental income growth compare to the trend in sales prices?
print("\nTrend Comparison:")
print("The overall trend in rental income growth and sales prices is as follows:")
print(average_growth_rates)

# Does this same trend hold true for all the neighborhoods across San Francisco?
print("\nNeighborhood Trends:")
# Identify neighborhoods with the highest average growth rates
top_neighborhoods = neighborhood_trends.groupby('neighborhood').mean().nlargest(5, 'rental_income_growth')
print("Top neighborhoods by average rental income growth:")
print(top_neighborhoods[['rental_income_growth', 'sales_price_growth']])
# Calculate the overall average growth rates across all neighborhoods
overall_growth = neighborhood_trends[['rental_income_growth', 'sales_price_growth']].mean()

print("\nOverall average growth rates across all neighborhoods:")
print(overall_growth)


# What insights can you share about the potential one-click, buy-and-rent strategy?
print("\nInvestment Strategy Insights:")
# Identify neighborhoods with the highest average growth rates in rental income and sales prices
highest_rental_income_growth_neighborhood = average_growth_rates['rental_income_growth'].idxmax()
highest_sales_price_growth_neighborhood = average_growth_rates['sales_price_growth'].idxmax()

print(f"The neighborhood with the highest average rental income growth is: {highest_rental_income_growth_neighborhood}")
print(f"The neighborhood with the highest average sales price growth is: {highest_sales_price_growth_neighborhood}")

print("These neighborhoods might be particularly attractive for the one-click, buy-and-rent strategy due to their strong growth trends.")
```
